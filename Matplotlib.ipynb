```python
#imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defining an example set 
x = np.array([1,2,3,4,5,6,7])
y = np.array([10,5,7,6,11,6,7])
```

```python
#to plot the scatter plot
plt.scatter(x,y, marker='o', c='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('X vs Y')
plt.show()
```

```python
#for a line plot
plt.plot(x,y,'--')
```

```python
#for a line of the form y=mx+c (y=1+x)
y = 1+x
plt.plot(x, y, '--')
```
